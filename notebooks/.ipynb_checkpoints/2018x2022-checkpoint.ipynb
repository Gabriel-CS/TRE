```python
import pandas as pd
import os
import numpy as np
from datetime import datetime
```

```python
urnas_2018 = pd.read_csv('../data/datasets/urnas_2018.csv', sep=';', encoding='utf-8-sig')
urnas_2022 = pd.read_csv('../data/datasets/urnas_2022.csv', sep=';', encoding='utf-8-sig')

chaves_cruzamento = ['CD_MUNICIPIO', 'NM_MUNICIPIO', 'NR_ZONA', 'NR_SECAO', 'TURNO']

# Converter as chaves para string para evitar erros de tipo
for df in [urnas_2018, urnas_2022]:
    for col in chaves_cruzamento:
        df[col] = df[col].astype(str)

# Fazer o cruzamento (merge)
# O parâmetro 'suffixes' vai renomear as métricas para sabermos de qual ano vieram
urnas = urnas_2018.merge(
    urnas_2022,
    on=chaves_cruzamento,
    how='inner',
    suffixes=('_2018', '_2022')
)

# 5. Criar o dataset apenas com as urnas que mantiveram a MESMA situação em ambos os anos
recorrentes = urnas.query("STATUS_2018 == STATUS_2022").copy()
recorrentes = recorrentes.query("STATUS_2018 != 0")

display(recorrentes[['NM_MUNICIPIO', 'NR_ZONA', 'NR_SECAO', 'STATUS_2018', 'STATUS_2022']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# 1. Prepara os dados
top10 = recorrentes['NM_MUNICIPIO'].value_counts().head(10).reset_index()
top10.columns = ['Município', 'Qtd_Urnas_Recorrentes']

# 2. Configura o tamanho
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

# 3. Desenha o gráfico
grafico = sns.barplot(
    x='Qtd_Urnas_Recorrentes', 
    y='Município',
    hue='Município',
    data=top10, 
    palette='Blues_r',
    legend=False
)

# 4. Adiciona os números nas barras
for index, value in enumerate(top10['Qtd_Urnas_Recorrentes']):
    grafico.text(value + 0.5, index, str(value), color='black', va='center', fontweight='bold')

# 5. Títulos e eixos
plt.title('TOP 10 Municípios com Mais Seções Recorrentes | 2018 x 2022', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Quantidade de Seções', fontsize=12)
plt.ylabel('Município', fontsize=12)
plt.xlim(0, top10['Qtd_Urnas_Recorrentes'].max() * 1.15)

plt.tight_layout()
plt.show()
```
